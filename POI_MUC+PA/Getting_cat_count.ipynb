```python
import pandas as pd
```

```python
MUC_df = pd.read_csv('MUC_poi_df.csv')
PA_df = pd.read_csv('PA_poi_df.csv')
```

```python
PA_compare_dict = { 
                'h3_code': PA_df.h3_source, 
                'total_poi': None,
                'station_count': None, 
                'Eat_and_Drink': None, 
                'Going_Out-Entertainment': None, 
                'Sights_and_Museums': None, 
                'Transport': None,
                'Accommodations': None, 
                'Shopping': None, 
                'Business_and_Services': None,
                'Facilities': None, 
                'Areas_and_Buildings': None}
PA_compare_df = pd.DataFrame(PA_compare_dict)
```

```python
PA_compare_df = PA_compare_df.drop_duplicates()
```

```python
def PA_stations_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_name == 'EV Charging Station')].count()
def PA_count_100_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Eat & Drink')].count()
def PA_count_200_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Going Out-Entertainment')].count()
def PA_count_300_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Sights and Museums')].count()
def PA_count_400_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Transport')].count()
def PA_count_500_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Accommodations')].count()
def PA_count_600_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Shopping')].count()
def PA_count_700_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Business and Services')].count()
def PA_count_800_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Facilities')].count()
def PA_count_900_in_h3(row):
    return PA_df[(PA_df.h3_source == row['h3_code']) & (PA_df.cat_definition == 'Areas and Buildings')].count()
def PA_total_poi(row):
    return row['Eat_and_Drink']+row['Going_Out-Entertainment']+row['Sights_and_Museums']+row['Transport']+row['Accommodations']+row['Shopping']+row['Business_and_Services']+row['Facilities']+row['Areas_and_Buildings']
def PA_area_high_low(row):
    if row['total_poi']<30:
        return 'Low-area'
    elif row['total_poi'] >= 30:
        return 'High-area'   
```

```python
PA_compare_df['station_count'] = PA_compare_df.apply(lambda row : PA_stations_in_h3(row), axis=1)
PA_compare_df['Eat_and_Drink'] = PA_compare_df.apply(lambda row : PA_count_100_in_h3(row), axis=1)
PA_compare_df['Going_Out-Entertainment'] = PA_compare_df.apply(lambda row : PA_count_200_in_h3(row), axis=1)
PA_compare_df['Sights_and_Museums'] = PA_compare_df.apply(lambda row : PA_count_300_in_h3(row), axis=1)
PA_compare_df['Transport'] = PA_compare_df.apply(lambda row : PA_count_400_in_h3(row), axis=1)
PA_compare_df['Accommodations'] = PA_compare_df.apply(lambda row : PA_count_500_in_h3(row), axis=1)
PA_compare_df['Shopping'] = PA_compare_df.apply(lambda row : PA_count_600_in_h3(row), axis=1)
PA_compare_df['Business_and_Services'] = PA_compare_df.apply(lambda row : PA_count_700_in_h3(row), axis=1)
PA_compare_df['Facilities'] = PA_compare_df.apply(lambda row : PA_count_800_in_h3(row), axis=1)
PA_compare_df['Areas_and_Buildings'] = PA_compare_df.apply(lambda row : PA_count_900_in_h3(row), axis=1)
PA_compare_df['total_poi'] = PA_compare_df.apply(lambda row : PA_total_poi(row), axis=1)
PA_compare_df['h3_area'] = PA_compare_df.apply(lambda row : PA_area_high_low(row), axis=1)
```
